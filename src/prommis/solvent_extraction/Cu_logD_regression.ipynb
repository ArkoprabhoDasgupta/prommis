```python
n_data = df.shape[0]

from math import log10
from idaes.core.util import model_statistics as mstat
import pyomo.environ as pyo
import numpy as np

m = pyo.ConcreteModel()

m.I = pyo.RangeSet(0, n_data-1)

# Input data vars
m.pH = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.C_ex = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CMn_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CCo_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CNi_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CCu_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)

# Output variable
m.logD = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)


# Parameters
n_a = 6
n_b = 2
m.a_set = pyo.Set(initialize = np.arange(1, n_a+1))
m.b_set = pyo.Set(initialize = np.arange(1, n_b+1))


m.ai = pyo.Var(m.a_set, domain=pyo.Reals, initialize=0.0)
m.bi = pyo.Var(m.b_set, domain=pyo.Reals, initialize=0.0)


name_mapper = {
    'pH_scaled': m.pH,
    '[D2EHPA] _scaled': m.C_ex,
    '[Mn] (mol/L)_scaled': m.CMn_fd,
    '[Co] (mol/L)_scaled': m.CCo_fd,
    '[Ni] (mol/L)_scaled': m.CNi_fd,
    '[Cu] (mol/L)_scaled': m.CCu_fd,
    'log D (obs)_scaled': m.logD
}

# Initialize input variables
for col in name_mapper.keys():
    for i in df.index:
        name_mapper[col][i].fix(df[col][i])

# Model variales

m.logD_scaled_pred =  pyo.Var(m.I, domain=pyo.Reals, initialize=1e-3)
m.Cext_net_pred = pyo.Var(m.I, domain=pyo.Reals, initialize=1e-3, bounds = (1e-3, 10))
m.logCext_net_pred = pyo.Var(m.I, domain=pyo.Reals, initialize=1e-3)

m.a = pyo.Var(m.I, domain=pyo.Reals, initialize=1e-3)
m.b = pyo.Var(m.I, domain=pyo.Reals, initialize=1e-3)


for i in range(3,7):
    m.ai[i].bounds = (0, None)

# Constraints

@m.Constraint(m.I)
def logCext_net_pred_constraint(m, i):
    return 10**(m.logCext_net_pred[i]) == (m.Cext_net_pred[i])

@m.Constraint(m.I)
def a_const(m, i):
    return m.a[i] == m.ai[1]  + m.ai[2]*m.Cext_net_pred[i]

@m.Constraint(m.I)
def b_const(m, i):
    return m.b[i] == m.bi[1] + m.bi[2]*m.logCext_net_pred[i]

@m.Constraint(m.I)
def Cext_net_pred_constraint(m, i):
    return m.Cext_net_pred[i] == m.C_ex[i] - m.ai[3]*m.CMn_fd[i] - m.ai[4]*m.CCo_fd[i]  -m.ai[5]*m.CCu_fd[i] - m.ai[6]*m.CNi_fd[i]

@m.Constraint(m.I)
def logD_scaled_pred_constraint(m, i):
    return m.logD_scaled_pred[i] == m.a[i]*m.pH[i] + m.b[i]
@m.Objective(sense = pyo.minimize)
def objective_function(m):
    return sum((m.logD_scaled_pred[i] - m.logD[i])**2 for i in m.I)

assert(mstat.degrees_of_freedom(m) == 8)

solver = pyo.SolverFactory('ipopt_v2')
solver.options['halt_on_ampl_error'] = 'yes'

results = solver.solve(m, tee=True, symbolic_solver_labels=True)
```

```python
n_data = df.shape[0]

from idaes.core.util import model_statistics as mstat
import pyomo.environ as pyo
import numpy as np

m = pyo.ConcreteModel()

m.I = pyo.RangeSet(0, n_data-1)

# Input data vars
m.pH = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.C_ex = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CMn_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CCo_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CNi_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.CCu_fd = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)

# Output variable
m.logD = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)


# Parameters
n_a = 6
n_b = 2
m.a_set = pyo.Set(initialize = np.arange(1, n_a+1))
m.b_set = pyo.Set(initialize = np.arange(1, n_b+1))


m.ai = pyo.Var(m.a_set, domain=pyo.Reals, initialize=0.0)
m.bi = pyo.Var(m.b_set, domain=pyo.Reals, initialize=0.0)


name_mapper = {
    'pH_scaled': m.pH,
    '[D2EHPA] _scaled': m.C_ex,
    '[Mn] (mol/L)_scaled': m.CMn_fd,
    '[Co] (mol/L)_scaled': m.CCo_fd,
    '[Ni] (mol/L)_scaled': m.CNi_fd,
    '[Cu] (mol/L)_scaled': m.CCu_fd,
    'log D (obs)_scaled': m.logD
}

# Initialize input variables
for col in name_mapper.keys():
    for i in df.index:
        name_mapper[col][i].fix(df[col][i])

# Model variales

m.logD_scaled_pred =  pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.Cext_net_pred = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.logCext_net_pred = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)

m.a = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)
m.b = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)

# Residual variable
m.res = pyo.Var(m.I, domain=pyo.Reals, initialize=0.0)


for i in range(3,7):
    m.ai[i].bounds = (0, None)

# Constraints

@m.Constraint(m.I)
def logCext_net_pred_constraint(m, i):
    return 10**(m.logCext_net_pred[i]) == m.Cext_net_pred[i]

@m.Constraint(m.I)
def a_const(m, i):
    return m.a[i] == m.ai[1]  + m.res[i]#m.ai[2]*m.Cext_net_pred[i]

@m.Constraint(m.I)
def b_const(m, i):
    return m.b[i] == m.bi[1] + m.bi[2]*m.logCext_net_pred[i]

@m.Constraint(m.I)
def Cext_net_pred_constraint(m, i):
    return m.Cext_net_pred[i] == m.C_ex[i] - m.ai[3]*m.CMn_fd[i] - m.ai[4]*m.CCo_fd[i]  -m.ai[5]*m.CCu_fd[i] - m.ai[6]*m.CNi_fd[i]

@m.Constraint(m.I)
def logD_scaled_pred_constraint(m, i):
    return m.logD_scaled_pred[i] == m.a[i]*m.pH[i] + m.b[i]
@m.Objective(sense = pyo.minimize)
def objective_function(m):
    return sum((m.logD_scaled_pred[i] - m.logD[i])**2 for i in m.I)

print(mstat.degrees_of_freedom(m))

solver = pyo.SolverFactory('ipopt_v2')
solver.options['halt_on_ampl_error'] = 'no'

results = solver.solve(m, tee=True, symbolic_solver_labels=True)
```
